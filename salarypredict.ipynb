```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Load the dataset
df = pd.read_csv("/content/salary.csv")

# Display the DataFrame to inspect the data
print(df)

# Extract features (X) and target variable (y)
X = df[["year of exp"]]
y = df["salary"]

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Display the training and testing sets
print("X_train:\n", X_train)
print("X_test:\n", X_test)
print("y_train:\n", y_train)
print("y_test:\n", y_test)

# Create a linear regression model
reg = LinearRegression()

# Fit the model to the training data
reg.fit(X_train, y_train)

# Make predictions on the test set
y_pred = reg.predict(X_test)

# Evaluate the model
score = reg.score(X_test, y_test)
print("R-squared score:", score)

# Plot the regression line along with the data points
plt.scatter(X, y, color='black', label='Data points')
plt.plot(X_test, y_pred, color='blue', linewidth=3, label='Regression line')
plt.xlabel('Years of Experience')
plt.ylabel('Salary')
plt.title('Linear Regression: Salary Prediction')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression

# Load your dataset (replace "/content/salary.csv" with the correct file path)
df = pd.read_csv("/content/salary.csv")

# Extract features (X) and target variable (y)
X = df[["year of exp"]]
y = df["salary"]

# Create and train the Linear Regression model
reg = LinearRegression()
reg.fit(X, y)

def predict_salary_from_input(model):
    try:
        # Get user input for years of experience
        years_of_experience = float(input("Enter the number of years of experience: "))

        # Use the model to predict the salary
        predicted_salary = model.predict([[years_of_experience]])

        print(f"For {years_of_experience} years of experience, the predicted salary is: {predicted_salary[0]:.2f}")

    except ValueError:
        print("Invalid input. Please enter a valid number.")

# Call the function with the trained model
predict_salary_from_input(reg)
```
